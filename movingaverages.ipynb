```python
import pandas as pd
import os
os.chdir("/Users/marcosjohnsonnoya/Desktop")
import csv
reader = pd.read_csv(r"IBM.csv")
def maaf(moving_average_length):
    total_days = (len(reader)-1)
    adjusteddays = total_days - moving_average_length
    maa = []
    i = 0
    while i < adjusteddays:
        maa.append(sum(reader.loc[(0+i):(moving_average_length-1+i),'Close'])/moving_average_length)
        i += 1
    mal = str(moving_average_length)
    maltot = mal + " Day Moving Average"
    delvals = list(range(adjusteddays,(total_days+1)))
    for x in delvals:
        reader.drop(x,inplace = True)
    reader.insert(7,maltot,maa,allow_duplicates=False) 
maaf(9)
maaf(200)
maaf(40)
print(reader)
```
